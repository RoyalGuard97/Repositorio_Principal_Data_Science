1) Cargar el dataset "Emisiones_CO2.csv" provisto en la clase 2 en un Dataframe de Pandas, quitar los registros que contengan valores faltantes e implementar una nueva columna, que contenga el resultado de una función Hash aplicada sobre el campo "Código de País" y se denomine "Clave_Hash".<br>
Consideraciones: Se puede utilizar la función provista.


```python
import hashlib
def hash_function(item):
    #Convertimos nuestro item un un tipo bytes, de codificacion ascii
    item = bytes(item,encoding="ascii")
    #Pasamos nuestro item por una funcion hash nativa de python
    item = hashlib.sha224(item).hexdigest()
    #De una cadena más larga e incomprensible construimos nuestro Hash Final. 
    return sum((index) * ord(character) for index, character in enumerate(repr(item), start=4))
```

```python

#Abrimos con pandas y lo codificamos (casteamos) a nuestro gusto
import pandas as pd
df = pd.read_csv("Emisiones_CO2.csv",sep="|",decimal=",",thousands=".",encoding="latin-1")
```

```python
#OJO eneralmente debemos preservar nuestro trabajo original.
#En este caso no es así, pero vemos que eliminamos los datos faltantes
#Mientras que inplace modifica la tabla original.
#Elimina todo registro que contengan datos nulos.
df.dropna(inplace=True)
```

```python
#Eliminando los Datos Faltantes
df.info()
```

```python
#Primero cambiaremos los nombres de nuestra tabla
df.rename(columns={"Código de país": "cod_pais", "Nombre del país":"nom_pais","Región": "region","Año":"anio","CO2 (kt)":"co2","CO2 per cápita (toneladas métricas)": "co2_percapita"},inplace=True)
```

```python
#Agregando la Columna de la Funcion Hash
#Se puede ver como una columna que se crea en base a otra. 
df["clave_hash"] = df["cod_pais"].apply(hash_function) #Sirve para aplicar una función a los datos de una columna. 
#Apply aplica una funcion a una columna para convertirla otra y agregarla
df
```

2. A partir del Dataframe creado en el punto 1, construir uno nuevo, que contenga solo los valores distintos de la tupla "Clave_Hash", "Código de País" , "Nombre de país" y "Región". Quitando luego del dataframe original los campos "Código de País" , "Nombre de país" y "Región"

```python
#Creo otro dataframe limitando las columnas del original
df_valores_unicos = df[["clave_hash","cod_pais","nom_pais","region"]]
#Me deshago de las tuplas, deshaciendo de los duplicados, y reseteo el indice dropeandolo facilmente
df_valores_unicos=  df_valores_unicos.drop_duplicates().reset_index(drop = True)
```

```python
#Una clave Hash por definición debe ser unica por ello la comprobamos.
df_valores_unicos["clave_hash"].is_unique
```

```python
#De la tabla original dropeamos las columnas que nos piden. axis = 1 para dropear columnas.tambien podríamos dropear registros, "recuerda."
df.drop(["cod_pais","nom_pais","region"],inplace = True,axis=1)
```
