# Relation

* Relación entre tipos de variables numéricas

```python
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset("tips")
tips.head(2)
```

## Relación Total Cuenta vs Propina

* Podremos ver que a mayor cuenta a pagar, mayor es la propina correspondiente

```python
sns.scatterplot(data=tips, x="total_bill",y="tip")
plt.show()
```


* Segmetación por Día

* Existen días donde se gasta mucho más y por ende se deja más propina

```python
sns.scatterplot(data=tips, x="total_bill",y="tip",hue="day")
plt.show()
```

## Cambiar Estilo de Puntos

* Los Puntos serán Almuerzo y las aspas representan las cenas. 

```python
sns.scatterplot(data=tips, x="total_bill",y="tip",hue="day",style="time")
plt.show()
```

## Manipular el tamaño de los puntos

* Podemos manipular el tamaño de los puntos según una escala numérica

```python
#Size es la cantidad de comensales.
sns.scatterplot(data=tips, x="total_bill",y="tip",hue="day",style="time",size="size")
plt.show()
```

## Mejorar la ubicación de la leyenda

* Con esto separamos la leyenda del entorno mero del gráfico

```python
sns.scatterplot(data=tips, x="total_bill",y="tip",hue="day",style="time",size="size")
plt.legend(loc="center",bbox_to_anchor=(1.12,0.5))
plt.show()
```

## Modificar Marcadores

```python
#Ya no son puntos, sino diamantes o cuadrados, diamond or square
markers = {"Lunch":"D","Dinner":"s"}
sns.scatterplot(data=tips, x="total_bill",y="tip",hue="day",style="time",size="size",markers=markers)
plt.legend(loc="center",bbox_to_anchor=(1.12,0.5))
plt.show()
```

## Modficar el Tamaño de un Gráfico Seaborn

```python
#Ya no son puntos, sino diamantes o cuadrados, diamond or square
plt.figure(figsize=(10,8))
markers = {"Lunch":"D","Dinner":"s"}
sns.scatterplot(data=tips, x="total_bill",y="tip",hue="day",style="time",size="size",markers=markers)
plt.legend(loc="center",bbox_to_anchor=(1.07,0.5))
plt.show()
```

## Gráficos Línea de Relación

```python
plt.figure(figsize=(10,8))
sns.lineplot(data=tips,x="total_bill",y="tip",hue="day",style="time",size="size")
plt.legend(loc="center",bbox_to_anchor=(1.07,0.5))
plt.show()
```

## Seaborn Relplot

* Funciona como el displot, se adapta según el contexto de tus datos y puedes modificar su tipo, con "kind"

```python
plt.figure(figsize=(1,8))
#Lo cambio con "line"
sns.relplot(data=tips,x="total_bill",y="tip",hue="day",style="time",size="size")
plt.show()
```

```python
plt.figure(figsize=(1,8))
#Lo cambio con "line"
sns.relplot(data=tips,x="total_bill",y="tip",hue="day",style="time",size="size",kind="line")
plt.show()
```

# Categorizar Gráficos por Columna

* Sin Estilo

```python
plt.figure(figsize=(1,8))
#Lo cambio con "line"
sns.relplot(data=tips,x="total_bill",y="tip",hue="day",style="time",size="size",col="time")
plt.show()
```

* Con estilo

```python
plt.figure(figsize=(1,8))
#Lo cambio con "line"
sns.relplot(data=tips,x="total_bill",y="tip",hue="day",size="size",col="time")
plt.show()
```
