# Subplots

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x ** 2
```

# Crear varios gr√°ficos

```python
#Primer gr√°fico de indice 1 - final
plt.subplot(1,2,1)
plt.plot(x,y)
#Segundo gr√°fico de indice 2 - final
plt.subplot(1,2,2)
plt.hist(y)
plt.show()
```

* Puedes representar mucho m√°s en una sola gr√°fica

```python
#Primer gr√°fico de indice 1 - final
plt.subplot(1,2,1)
plt.plot(x,y,"r--")
plt.plot(y,x,"b:")
#Segundo gr√°fico de indice 2 - final
plt.subplot(1,2,2)
plt.pie(y)
plt.show()
```

* El primer indice del subplot, hace referencia a una fila, el segundo hace referencia a una columna

```python
#Primer gr√°fico de indice 1 - final
plt.subplot(2,1,1)
plt.plot(x,y,"r--")
plt.plot(y,x,"b:")
#Segundo gr√°fico de indice 2 - final
plt.subplot(2,1,2)
plt.pie(y)
plt.show()
```
