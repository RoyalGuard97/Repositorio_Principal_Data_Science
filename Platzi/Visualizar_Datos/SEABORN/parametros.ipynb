# Manejo de Parametros en Seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
tips = sns.load_dataset("tips")
```

* Este es un dataframe precargado

```python
tips
```

## Gráfico dissplot

* Realizamos una Gráfica de distribución sobre el "total_cuenta"

```python
#Para esta gráfica requerimos el set de datos y es aspecto a evaluar. 
#En otras palabras la variable 
sns.set(style="whitegrid",palette="dark",font="Calibri",font_scale=1.25)
sns.displot(data = tips, x="total_bill")
plt.xlabel("Total de Cuenta")
plt.ylabel("Frecuencia")
plt.title("Distribución de Total de Cuenta")
plt.show()
```

## Graficar Relaciones de 2 Variables

* Parecido a un Scatterplot

```python
sns.set(style="whitegrid",palette="dark",font="Calibri",font_scale=1.25)
sns.displot(data = tips, x="total_bill",y="tip")
plt.xlabel("Total de Cuenta")
plt.ylabel("Total de Propinas")
plt.title("Cuenta vs Propinas (Total)")
plt.show()
```

```python
sns.set(style="whitegrid",palette="dark",font="Calibri",font_scale=1.25)
sns.scatterplot(data = tips, x="total_bill",y="tip")
plt.xlabel("Total de Cuenta")
plt.ylabel("Total de Propinas")
plt.title("Cuenta vs Propinas (Total)")
plt.show()
```

## Como Agrupar?

* Esto se hace con el apartado "hue"
* Agruparemos por ejemplo por sexo

```python
sns.set(style="whitegrid",palette="Spectral_r",font="Calibri",font_scale=1.25)
sns.scatterplot(data = tips, x="total_bill",y="tip",hue="sex")
plt.xlabel("Total de Cuenta")
plt.ylabel("Total de Propinas")
plt.title("Cuenta vs Propinas (Total)")
plt.show()
```

## También funciona en Histogramas

```python
#Para esta gráfica requerimos el set de datos y es aspecto a evaluar. 
#En otras palabras la variable 
sns.set(style="whitegrid",palette="dark",font="Calibri",font_scale=1.25)
sns.displot(data = tips, x="total_bill",hue="sex")
plt.xlabel("Total de Cuenta")
plt.ylabel("Frecuencia")
plt.title("Distribución de Total de Cuenta")
plt.show()
```

## Cambiando el tipo de Gráfico

* Haremos del Histograma un gráfico de Densidad

```python
#Para esta gráfica requerimos el set de datos y es aspecto a evaluar. 
#En otras palabras la variable 
sns.set(style="whitegrid",palette="dark",font="Calibri",font_scale=1.25)
sns.displot(data = tips, x="total_bill",hue="sex",kind="kde")
plt.xlabel("Total de Cuenta")
plt.ylabel("Frecuencia")
plt.title("Distribución de Total de Cuenta")
plt.show()
```

De: sns.displot(data = tips, x="total_bill",hue="sex",kind="kde")
