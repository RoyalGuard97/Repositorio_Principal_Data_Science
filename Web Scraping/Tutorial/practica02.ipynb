# Listas de Monedas 

```python
nom_monedas_sucio = [] 
```

# Extracci√≥n de Datos

```python
contenido = driver.page_source
```

```python
contenido
```

```python
soup = bs(contenido)
```

```python
soup
```

```python
for tarjeta in soup.find_all("div",attrs={"class":"homepageWidget cryptoHP"}):
    nom_moneda = tarjeta.find("tbody")
    nom_monedas_sucio.append(nom_moneda.text)
    
```

```python
nom_monedas_sucio
```

```python
cadena_sucia = nom_monedas_sucio[0]
```

```python
cadena_sucia
```

```python
cadena_sucia.replace("\n","")
```

```python
lista = cadena_sucia.split("\n")
```

```python
lista
```

```python
df = pd.DataFrame(lista)
```

```python
df.columns = ["columna"]
```

```python
condicion = df["columna"] != ""
```
