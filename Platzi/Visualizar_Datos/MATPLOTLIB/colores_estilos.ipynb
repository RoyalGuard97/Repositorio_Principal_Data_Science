# MOD MATLAB

```python
#Fig es el entorno y axis es el gráfico propiamente dicho. 
plt.style.use("dark_background")
fig,axis = plt.subplots(figsize=(8,8))
axis.plot(x,x+1,"o--")
axis.plot(x,x+2,"r-")
axis.plot(x,x+3,"y.-")
axis.plot(x,x+4,"go:")
fig.set_facecolor("gray")
```

# MOD PYPLOT

* El cambio es en el color

```python
#Fig es el entorno y axis es el gráfico propiamente dicho. 
plt.style.use("dark_background")
fig,axis = plt.subplots(figsize=(8,8))
axis.plot(x,x+1,color="green",alpha=0.5,) #alpha es opacidad. 
axis.plot(x,x+2,color = "blue")
axis.plot(x,x+3,color="#66FF66")
axis.plot(x,x+4,color="#CCFFCC",linewidth=10) #Grosor de la línea. 
#Otros estilos se pueden aplicar con "linestyle o marker and markersize"
#Incluso puedes cambiar el color del marcador con markerfacecolor 
plt.show()
```
