# Sistemas de Ecuaciones Especiales

* y = 3X + 5
* y = -1x + 3
* y = 2x +1

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(-6,6)
y_1 = 3*x + 5
y_2 = -x + 3
y_3 = 2*x +1
```

```python
plt.figure()
plt.plot(x,y_1)
plt.plot(x,y_2)
plt.plot(x,y_3)

plt.xlim(-8,8)
plt.ylim(-8,8)

plt.axvline(x=0,color="grey")
plt.axhline(y=0,color="grey")
```

* Se observa que no existe solución factible para el sistema
* Sistema Sobredeterminado

* Sistema de Ecuación con Solución

```python
x = np.arange(-6,6)
y_2 = -x + 3
y_3 = 2*x + 1
```

```python
plt.figure()
plt.plot(x,y_2)
plt.plot(x,y_3)

plt.xlim(-8,8)
plt.ylim(-8,8)

plt.axvline(x=0,color="grey")
plt.axhline(y=0,color="grey")
```

* Sistema de Ecuaciones de Infinitas Soluciones

* Las Rectas son iguales por ende hay infinitas soluciones

```python
x = np.arange(-6,6)
y_2 = -2*x - 1
y_3 = 2*x + 1
```

```python
plt.figure()
plt.plot(x,y_2)
plt.plot(x,-y_3)

plt.xlim(-8,8)
plt.ylim(-8,8)

plt.axvline(x=0,color="grey")
plt.axhline(y=0,color="grey")
```
