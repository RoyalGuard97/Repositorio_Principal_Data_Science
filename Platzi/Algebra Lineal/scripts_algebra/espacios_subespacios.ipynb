# Espacios & SubEspacios

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
v1 = np.array([1,1])
v2 = np.array([-1,-1])
```

```python
for a in range(-10,10):
    for b in range(-10,10):
        plt.scatter(v1[0]*a + v2[0]*b,v1[1]*a + v2[1]*b, marker=".",color="orange")

plt.xlim(-20,20)
plt.ylim(-20,20)

plt.axvline(x=0,color="grey")
plt.axhline(y=0,color="grey")
```

* Con dos vectores de este tipo, creamos solo una recta.
* Los Espacios o SubEspacios dependen de esos mismos vectores

```python
v1 = np.array([1,0])
v2 = np.array([2,-3])
```

```python
for a in range(-10,10):
    for b in range(-10,10):
        plt.scatter(v1[0]*a + v2[0]*b,v1[1]*a + v2[1]*b, marker=".",color="orange")

plt.xlim(-20,20)
plt.ylim(-20,20)

plt.axvline(x=0,color="grey")
plt.axhline(y=0,color="grey")
```

* En este caso si creamos un espacio 2D donde podemos obtener puntos en funciones de estos

```python
from mpl_toolkits.mplot3d import Axes3D
v1 = np.array([1,0,0])
v2 = np.array([2,-3,0])

fig = plt.figure()
ax = fig.add_subplot(111,projection="3d")

for a in range(-10,10):
    for b in range(-10,10):
        ax.scatter(v1[0]*a + v2[0]*b,v1[1]*a + 
        v2[1]*b,v1[2]*a+v2[2], 
        marker=".",color="orange")

ax.set_xlabel("Eje X")
ax.set_ylabel("Eje Y")
ax.set_zlabel("Eje Z")

ax.view_init(elev=10,azim=10)
plt.axvline(x=0,color="grey")
plt.axhline(y=0,color="grey")
```

* Un Hiperplano siempre es una dimensión menos de la cual actualmente nosotros estamos trabajando.
