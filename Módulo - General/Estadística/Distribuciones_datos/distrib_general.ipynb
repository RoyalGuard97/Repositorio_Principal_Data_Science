# Distribución de Datos e Histogramas

```python
import matplotlib.pyplot as plt
datos = list(range(1,11))
```

* datos: la data a analizar
* color: color de las barras
* ec: remarca individualmente las barras

```python
#Gráfica de Frecuencias en Histograma
plt.hist(datos, color="pink",ec="black")
plt.show()
```

```python
datos = [1,1,1,2,2,3,4,5,5,6]
```

```python
#Gráfica de Frecuencias en Histrogramas
plt.hist(datos,color="red",ec="black")
plt.show()
```

# Generar arreglos con datos aleatorios

```python
import numpy as np
datos = list(np.random.rand(10))
```

```python
#Gráfica de Frecuencias en Histrogramas
plt.hist(datos,color="red",ec="black")
plt.show()
```

* Se construye una distribución normal con media igual a 100 (loc)
* Se construye una distribución normal con desviación igual a 10 (scale)
* El tamaño es de 100 datos

```python
#Distribución Normal en Numpy
datos = np.random.normal(loc = 100, scale=10, size = 1000)
```

```python
#Gráfica de Frecuencias en Histrogramas
plt.hist(datos,color="red",ec="black")
plt.show()
```

# Normal, Uniforme y Exponencial

```python
#Distribución Normal
datos = np.random.normal(size = 1000000)
```

```python
plt.hist(datos,color="pink",ec="black")
plt.title("Distribución Normal")
plt.show()
```

```python
#Distribución Uniforme
datos = np.random.uniform(size = 1000000)
```

```python
plt.hist(datos,color="pink",ec="black")
plt.title("Distribución Uniforme")
plt.show()
```

```python
#Distribución Exponencial
datos = np.random.exponential(size = 1000000)
```

```python
plt.hist(datos,color="pink",ec="black")
plt.title("Distribución Exponencial")
plt.show()
```
