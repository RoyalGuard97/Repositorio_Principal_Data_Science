# Distribuciones

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

## Bonus
* plt.savefig(“name.png”) → Guardar Imágenes de los Gráficos

```python
tips = sns.load_dataset("tips")
tips.head()
```

* Gráfico con Cola hacia la Derecha

```python
#Histograma de las "Propinas"
sns.histplot(data=tips, x = "tip",bins=15)
plt.show()
```

* Comportamiento Acumulativo

```python
sns.histplot(data=tips, x = "tip",bins=15,cumulative=True)
plt.show()
```

Histograma Avanzado

* Podemos observar la distribución de las Propinas y su clasificación por Sexo

```python
sns.histplot(data=tips, x="tip",bins=15,hue="sex",stat="count",multiple="dodge")
```

## Gráfico de Densidad

```python
#Cumulative
sns.kdeplot(data=tips, x="tip",hue="sex",shade=True)
plt.show()
```

```python
sns.ecdfplot(data=tips, x = "tip", hue="sex")
plt.show()
```

## Displot es como la gráfica general

```python
sns.displot(data=tips, x = "tip", hue="sex",kind="hist",multiple="stack")
plt.show()
```
