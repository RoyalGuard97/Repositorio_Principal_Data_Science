```python
df = pd.read_csv("Tipo de dato.txt",delim_whitespace=True)
```

```python
#deshacernos de las mayúsculas en las columnas
df.columns = df.columns.str.lower()
```

```python
#Guardar la última fila o registro
suma_totales = df.tail(1)
```

```python
#Dropeamos la última fila a través de su indice
df.drop(df.tail(1).index,inplace=True)
```

```python
df
```

```python
#Convertir una columna a tipo fecha. 
df["fecha"] = pd.to_datetime(df["fecha"])
```

```python
df["cuenta"] = df["cuenta"].astype(int)
```

```python
df["departamento"] = df["departamento"].astype(int)
```

```python
lista_suma_importe = [float(elemento.strip("(").strip(")").strip("$").replace(",","")) for elemento in df["suma_de_importe"]]
```

```python
df["suma_de_importe"] = lista_suma_importe
```
