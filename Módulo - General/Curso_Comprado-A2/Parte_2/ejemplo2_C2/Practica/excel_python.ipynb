# Operaciones con Pandas

```python
df_precios.info()
```

```python
def conversor (elemento):
    return elemento.strip().lstrip("$")
```

```python
df_precios["Precio"] = df_precios["Precio"].apply(conversor).astype(float)
df_precios["Costo"] = df_precios["Costo"].apply(conversor).astype(float)
```

```python
df_precios.head()
```

```python
df_precios["Ganancia"] = df_precios["Precio"] - df_precios["Costo"]
```

# Merging

* Juntar 2 Dataframes usando un Left Join
* Buscar V en Excel
* Ten cuidado con el formato que se utiliza en las columnas de unión
