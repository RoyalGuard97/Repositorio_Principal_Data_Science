# Etiquetas, tículos, leyendas

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = np.sin(x)
```

* ¿Qué intentas representar?

```python
fig,axis = plt.subplots(1,2,figsize=(8,5))
axis[0].plot(x,y,label="$Sin(x)$") #Le agregamos el nombre de lo graficado.
axis[0].set_title("Relacion X vs Y")
axis[0].set_xlabel("Los Valores de X")
axis[0].set_ylabel("Los Valores de Y")
axis[0].legend() #Es necesario aplicar leyenda para que aparezca. 
#-----------------Gráfico 2---------------------------
axis[1].plot(y,x)
axis[1].set_title("Relación Y vs X")
axis[1].set_xlabel("Los Valores de Y")
axis[1].set_ylabel("Los Valores de X")
fig.tight_layout() #Mejora la visualización - Diseño No apretado. 
plt.show()
```

# Método Alternativo más Simple

```python
plt.plot(x,y,label="sin(x)")
plt.title("Este es el Gráfico Seno de X")
plt.xlabel("Los valores de X")
plt.ylabel("Los valores de Y")
plt.legend(loc = "upper right") #El parametro loc es el de la localización. 
plt.show()
```
