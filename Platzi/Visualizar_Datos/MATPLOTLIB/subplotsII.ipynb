# Subplots Orientado a Objetos

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = np.sin(x)
```

```python
#Genero dos figuras, en una fila con dos columnas, puedo acceder a ellas como en una lista.
fig,axes = plt.subplots(1,2)
axes[0].plot(x,y,"b")
axes[1].plot(y,x,"r")
```

# Método alternativo - Tupla

```python
#Ventaja de trabajar con un método orientado a objetos. 
fig, (ax1,ax2) = plt.subplots(2,1) # Filas Columnas
ax1.plot(x,y,"b")
ax2.plot(y,x,"r")
plt.show()
```

```python
#Con figsize manipulamos el tamaño!!
fig, axis = plt.subplots(2,2,figsize=(5,5)) # Filas Columnas
axis[0,0].plot(x,np.sin(x),"rD:")
axis[0,1].hist(y,color="yellow")
axis[1,0].pie(x)
axis[1,1].boxplot(np.cos(y))
#Mejora la visualizacion
fig.tight_layout()
plt.show()
```
