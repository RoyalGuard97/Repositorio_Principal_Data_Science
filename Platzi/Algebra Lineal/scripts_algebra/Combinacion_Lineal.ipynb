# Combinaciones Lineales

```python
import numpy as np
import matplotlib.pyplot as plt
from funciones_aux import graficarVectores
```

```python
#Vectores que serán sometidos a una combinación lineal
v1 = np.array([2,5])
v2 = np.array([3,2])
```

```python
#Generamos el vector producto de la combinación Lineal. 
v1v2 = 2*v1 + 3*v2
```

* Tengamos en cuenta la magnitud del vector!!

* Hacemos referencia a la regla del paralelogramo

```python
#Empleamos la función de nuestro script "funciones_aux.py"
graficarVectores([v1,v2,v1v2],["orange","blue","red"])
#Determinamos los ejes visibles. 
plt.xlim(-1,17)
plt.ylim(-1,17)
```

* Sub-Cojunto de Combinaciones lineales

```python
for a in range(-10,10):
    for b in range(-10,10):
        plt.scatter(v1[0]*a + v2[0]*b,v1[1]*a + v2[1]*b, marker=".",color="orange")

plt.xlim(-100,100)
plt.ylim(-100,100)

plt.axvline(x=0,color="grey")
plt.axhline(y=0,color="grey")
```
