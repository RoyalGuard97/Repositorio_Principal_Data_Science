```python
import pandas as pd
df = pd.read_csv("hospitales2.csv")
```

```python
df.head()
df.shape
```

```python
#Si crear una tabla a partir de una vieja, procura someter a strip a la ultima columna
#Para manipular columnas en pandas lo mejor son las listas iterables. 
coordenadas = [elemento.lstrip("POINT (").rstrip(")").split() for elemento in df["WKT"]]
latitude = [float(elemento[1]) for elemento in coordenadas]
longitude = [float(elemento[0]) for elemento in coordenadas]
lista_telefono = [elemento.strip() for elemento in df["TELEFONO"]]
```

```python
#En caso de que los elementos pasaran solo a otro dataframe, podríamos aplicar esto de modo diferente.
dicc = dict(latitud = latitude,longitud = longitude,calle = df["CALLE"],telef = lista_telefono)
df_new = pd.DataFrame(dicc)
```
