# Distribución Binomial

```python
import matplotlib.pyplot as plt
import math
```

* Fórmula de distribución binomial

```python
def binomial (x,n,p):
    factoriales = math.factorial(n)/(math.factorial(x) * math.factorial(n-x))
    probabilidades = p**x * (1-p)**(n-x)
    return factoriales * probabilidades
```

```python
ns = [6,10,20,40,80,160]
p = 0.5

for n in ns:
    Xs = [k/n for k in range(n+1)]
    Ys = [binomial(x,n,p) for x in range(n+1)]
    plt.plot(Xs,Ys,label=f"n = {n}")

plt.xlabel("Probabilidades")
plt.ylabel("$f(x)$")
plt.legend()
plt.show()
```
