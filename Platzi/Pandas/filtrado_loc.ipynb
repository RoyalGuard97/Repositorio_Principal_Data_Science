# LOC & ILOC

```python
import pandas as pd 
df = pd.read_csv("betsellers.csv")
```

```python
df.head(5)
```

* Slicing

```python
#Filtrado por Indice
df[0:4]
```

```python
#Para filtrar un dataframe por columna, se necesita una lista con su nombre.
df[["Name"]]
```

* LOC

```python
df.loc[:]
```

```python
#Con el loc, en un slicing el elemento final se incluye
df.loc[0:4]
```

* Busqueda Combinada

```python
#Hemos traido los registros de indice 100 a 105 (inclusivo) pero solo las columnas "Name" & User Rating
df.loc[100:105,["Name","User Rating"]]
```

```python
#Podemos afectar valores filtrados, solo como visualizaciÃ³n. 
df.loc[:,["Reviews","User Rating"]] /10 #No se altera el dataframe original. 
```

* Filtrado por Condicion con resultado booleano

```python
#Obtenemos que el primer registro de indice 0 cuenta con ello
df.loc[:,["Author"]] == "JJ Smith"
```

* ILOC

```python
df.iloc[:]
```

* ILOC es similar a Loc, pero solo se aceptan valores numÃ©ricos para las bÃºsquedas, osea todo lo hacemos mediante indices

```python
#Antes de la coma trabajas con filas, despuÃ©s de la coma trabajas con columnas
df.iloc[:,0:4]
```

```python
#Valor especÃ­fico - del segundo registro, muestrame la columna de indice 3
df.iloc[1,3]
```
