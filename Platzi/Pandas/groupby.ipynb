# Groupby en Pandas

* Usemos Funciones de Agregacion

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("betsellers.csv")
```

* Qué tan comunes son los registros por autor?

```python
df.groupby("Author").count().sort_values("Name",ascending=False)
```

* Funcion agregación de conteo

```python
df.groupby("Author").count()
```

* Función agregación Media

```python
df.groupby("Author").mean()
```

```python
df.groupby("Author").min()
```

```python
df.groupby("Author").max()
```

* Funcion Agregación de Suma

```python
df.groupby("Author").sum()
```

* Cuando el nombre de una columna se "baje" quiere decir que no es una columna normal, sino que funge como nuestro índice

```python
df.groupby("Author").min().loc["Adir Levy"]
```

* Si deseamos que los indices de nuestra agrupación sean los estándares de siempre, pues procedemos a reseteat los indices

```python
#Esto es muy sencillo de aplicar - recuerda que puedes usar "drop"
df.groupby("Author").sum().reset_index()
```

* Emplear varias funciones de agregación en un agrupamiento

```python
#Aplicamos dos métricas distintas
df.groupby("Author").agg(["min","max"])
```

## Agrupar empleando diccionarios

```python
#Con esto podemos limitar el resultado, y emplear funciones de agregación diferentes para cada columna a analizar. 
df.groupby("Author").agg({"Reviews":["min","max"], "User Rating": "sum"})
```

* Podemos emplear funciones lambda para obtener funciones customizadas en un agrupamiento

```python
df.groupby('Author').agg({'Year':lambda x: [i-2000 for i in x], 'Reviews':lambda x: sum([i**2 for i in x])})
```

## Crear Llaves compuestas

```python
#Agrupamos por autor y por año. 
df.groupby(["Author","Year"]).sum()
```
