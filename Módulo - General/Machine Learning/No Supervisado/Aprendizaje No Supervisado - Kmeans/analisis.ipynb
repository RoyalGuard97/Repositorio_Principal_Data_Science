# Para aprendizaje no Supervisado - Hallar Patrones

```python
#De la liberia sklearn.cluster
from sklearn.cluster import KMeans #KMeans sería el algoritmo a emplear
kmeans = KMeans(n_clusters=3).fit(df) #Nuestro algoritmo estare en base a 3 clusters. Y entrenamos con df
centroids = kmeans.cluster_centers_ #Así hallamos los centroides. 
print(centroids)
```

```python
plt.scatter(df["x"],df["y"], c=kmeans.labels_.astype(float),s=50,alpha=0.75)
plt.scatter(centroids[:,0],centroids[:,1],c="red",s=50)
plt.show()
```

```python
plt.scatter(df["x"],df["y"], c=kmeans.labels_,s=50,alpha=0.75)
plt.scatter(centroids[:,0],centroids[:,1],c="red",s=50)
plt.show()
```
