```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.random.randint(1,50,100)
```

# Gráficar la Distribución de los Datos

```python
plt.hist(data,bins=10)
plt.show()
```

```python
plt.hist(data,bins=10,histtype="step")
plt.show()
```

## Boxplot

```python
data = np.append(data,200)
plt.boxplot(data,vert=False,patch_artist=True, notch=True) #Vertical, o rellenar. 
#Ocultar Valores atípicos showfliers = False
plt.show()
```

## Scatter

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
area = (30*np.random.rand(N)) ** 2
colors = np.random.rand(N)
```

```python
plt.scatter(x,y,s=area,c=colors,marker="o",alpha=0.5) #Investigar lo del area
#Marker es el marcador puede ser punto o equis por ejemplo
#Alpha es para el aspecto de opacidad. 
plt.show()
```
