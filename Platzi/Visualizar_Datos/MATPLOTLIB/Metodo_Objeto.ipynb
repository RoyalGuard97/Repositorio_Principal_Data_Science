# Método Orientado a Objetos

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.5,0.9])
axes.plot(x,y,"b")
plt.show()
```

* Son objetos Distintos que reposan en una misma figura
* Mucho más personalizable

```python
fig = plt.figure()
#Manipulamos las Dimensiones del Gráfico
axes = fig.add_axes([0.1,0.1,0.8,0.9])
axes2 = fig.add_axes([0.2,0.55,0.4,0.3])
axes.plot(x,y,"b")
axes.set_facecolor("orange")
axes2.plot(y,x,"r")
axes2.set_facecolor("gray") #Cambia el color de fondo
plt.show()
```
